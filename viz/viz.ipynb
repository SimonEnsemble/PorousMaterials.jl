```python
import py3Dmol
```

```python
def read_xyz(xyz_filename):
    f = open(xyz_filename)
    lines = f.read()
    f.close()
    return lines

xyz = read_xyz("IRMOF-1.xyz")

def read_cube(cube_filename):
    f = open(cube_filename)
    lines = f.read()
    f.close()
    return lines

cube = read_cube("CH4_in_IRMOF-1.cube")

def read_vtk(vtk_filename):
    f = open(vtk_filename)
    lines = f.read()
    f.close()
    return lines

vtk = read_vtk("IRMOF-1.vtk")
```

```python
v = py3Dmol.view(width=400,height=400)
v.addModel(xyz, 'xyz')
v.setStyle({'stick':{}})
v.addVolumetricData(cube, "cube", {'isoval': 0.0, 'color': "yellow", 'opacity': 0.5})
# voldata = py3Dmol.VolumeData(cube, "cube")
# v.addIsosurface({'isoval': -10.0, 'color': "blue", 'opacity': 0.5})
v.addUnitCell()
v.zoomTo()
v.setBackgroundColor('0xeeeeee')
v.animate({'loop': 'backAndForth'})
v.addUnitCell(vtk, "vtk")
# v.setBackgroundColor('fffff0')
v.show()
```
